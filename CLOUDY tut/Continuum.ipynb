```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('bmh')
```

```python
pn = pd.read_csv('script.con', sep = "\t")
solar = pd.read_csv('solar_ab.con', sep = "\t")
blr = pd.read_csv('blr.con', sep = "\t")
pn_ovr = pn = pd.read_csv('script.ovr', sep = "\t")
pn_ovr.columns.values
```

```python
plt.figure(figsize = (14,5))
plt.loglog(pn['#Cont  nu'], pn['total'], label = 'planetary nebula')
plt.loglog(solar['#Cont  nu'], solar['total'], label = 'solar')
plt.xlabel(r'$\lambda$ [$\mu m$]')
plt.ylabel('flux')
plt.xlim(0.1,1000)
plt.ylim(1e33, 1e40)
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (14,5))
plt.loglog(pn['#Cont  nu'], pn['net trans'], label = 'planetary nebula')
plt.loglog(solar['#Cont  nu'], solar['net trans'], label = 'solar')
plt.xlabel(r'$\lambda$ [$\mu m$]')
plt.ylabel('flux')
plt.xlim(0.1,1000)
plt.ylim(1e33, 1e40)
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (14,5))
plt.loglog(blr['#Cont  nu'], blr['total'], label = 'BLR')
plt.xlabel(r'$\lambda$ [$\mu m$]')
plt.ylabel('flux')
plt.xlim(0.1,10)
plt.ylim(1e7, 1e11)
plt.legend()
plt.show()
```
